```python
j_entry = [
    {'meal' : 'Bananas and peanutbutter on toast', 'enjoyed' : True},
    {'meal' : 'yogurt with granola and fresh fruit', 'enjoyed' : True},
    {'meal' : 'protein shake with flax and chia seeds', 'enjoyed' : False},
    {'meal' : 'pre made salad from grocery store', 'enjoyed': False},
    {'meal' : 'fajita tacos with bae', 'enjoyed':True},
    {'meal' : 'chocolate lava cake with caramel ice cream', 'enjoyed':True},
    {'meal' : 'boba tea with the girls' , 'enjoyed': True}
]

def list_meals():
    for index, task in enumerate(j_entry):
        print(str.format('{}: {} (enjoyed: {})', index, task['meal'], task['enjoyed']))


def add_meal():
    meal_text = input('Add a new meal: ')
    new_meal= {'meal': meal_text, 'enjoyed': False}
    j_entry.append(new_meal)
    
def take_meal():
    list_meals()
    remove_meal= int(input('What meal do you want to delete?'))
    j_entry.remove(j_entry[remove_meal])

def enjoy_meal():
    list_meals()
    yum_meal=int(input("which meal was yummy?"))
    j_entry[yum_meal]['enjoyed']= True


    
    
menu_text = """
Welcome to your interactive meal journal, what are we doing today?
====================
1. Show meal history
2. Add a meal
3. Remove a meal entry
4. Mark food as enjoyed
5. Quit

What can I help you with today? """

program_is_running = True

while program_is_running:
    decision = input(menu_text)
    if decision == '1':
        list_meals()
    elif decision == '2':
        add_meal()
    elif decision== '3':
        take_meal()
    elif decision== '4':
        enjoy_meal()


    elif decision == '5':
        program_is_running = False
        print("Thank you for your entry, have a great day!")
    else:
        print('please choose a valid option')
```
