```python
#looking at the presence of absense of frogs across all sites, looks like the first near 100 sites had frogs while the rest did not
sns.catplot(x="pres.abs",y='Unnamed: 0',data=frogs)
```

```python
#checking the altitude of the check sites, 
#because there seems to be correlation between site number and presence of frogs, doesnt SEEM to be any correlation between presence and absense
sns.catplot(x="Unnamed: 0",y='altitude', hue= 'pres.abs', data=frogs)
```

```python
#also looking at the rain at certain altitudes and if it affects the presence of frogs 
# seems like frogs like mid altitudes and low rain
sns.relplot( x= "avrain", y="altitude", hue= "pres.abs", data=frogs )
```
